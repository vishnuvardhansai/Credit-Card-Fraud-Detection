Importing The Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
card_data = pd.read_csv('/content/creditcard.csv')
```

```python
card_data.head()
```

```python
card_data.tail()
```

```python
#Dataset Info

card_data.info()
```

```python
# checking missing values
card_data.isnull().sum()
```

```python
#Legit and Fraudulent Transactions
card_data['Class'].value_counts()
```

```python
#Legit Transactions
legit = card_data[card_data.Class == 0]
legit.shape
```

```python
fraud = card_data[card_data.Class == 1]
fraud.shape
```

```python
#Statistical Measures
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both Transactions
card_data.groupby('Class').mean()
```

```python
#Under Sampling
#Build a sample dataset conataining similar distribution of normal and fraud Transactions
legit_samples = legit.sample(n=56)
```

```python
#concate The Two Data Frames
dataset_new = pd.concat([legit_samples, fraud], axis=0) # row Wise(0)
```

```python
dataset_new.head()
```

```python
dataset_new.tail()
```

```python
dataset_new['Class'].value_counts()
```

```python
dataset_new.groupby('Class').mean()
```

```python
#Split The Data Into Features And Targets
X = dataset_new.drop(columns='Class', axis=1)
Y = dataset_new['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split The Data into Training and Testing Data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training
model_train = LogisticRegression() #Using Logistic Regression to train the model
```

```python
#Train The Logistic Regression Model  With Training set of Data
model_train.fit(X_train,Y_train)
```

```python
#Evaluating the Model
#Accuracy on Training Data
X_train_prediction = model_train.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data : ', train_data_accuracy)
```

```python
#Accuracy On Test Data
X_test_prediction = model_train.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test Data : ',test_data_accuracy)
```
